```python
import numpy as np
import pandas as pd #it uses to dataset
import difflib   # it helps when user searching some movie name mistakely,it will find the colost match some value
from sklearn.feature_extraction.text import TfidfVectorizer #it helps to change textual data into numerical data
from sklearn.metrics.pairwise import cosine_similarity    #when user give input , name of their movie and based on this data or 
                                                     #input,we are going to suggest movies they can watch
                                                        # it used to similar between two vectors

movies_data=pd.read_csv('movies.csv')
movies_data.head() #head mean it load data fast and it only shows first five 
```

```python
movies_data.shape #it shows shape of the data,how many columns and rows in their ,
                    #columns represent how many movies in there, and rows represent movies information name,genres,taglist
```

```python
movies_data.isnull().sum() #it is used to find missinng value is this dataset
```

```python
movies_data.info() #information about the dataset
```

```python
#selecting the relevant factors for recommendation 
```

```python
selected_features=['genres','keywords','title','cast','director','tagline']
selected_features
```

```python
#replace missing the null values 
for feature in selected_features:
    movies_data[feature]=movies_data[feature].fillna(' ') #i am fill with empty string
```

```python
movies_data.isnull().sum()
```

```python
#combining the text data to feature vectors
combined_features=movies_data['genres']+''+movies_data['keywords']+''+movies_data['title']+''+movies_data['cast']+''+movies_data['director']
combined_features
```

```python
#it convert textual data into numerical data
vectorizer=TfidfVectorizer()
feature_vector=vectorizer.fit_transform(combined_features)
print(feature_vector)
```

```python
#it shows similarity score it matches it shows 'one' it doesnot match, it shows 0,if it sligthy matches 0.020 some like this 
#it shows close match 
similarity=cosine_similarity(feature_vector)
similarity
```

```python
similarity.shape
```

```python
#getting the movie name from the user
movie_name=input('Enter the favourite movie name: ')
movie_name
```

```python
#creating a list with all movie name given in this datset
list_of_all_titles=movies_data['title'].tolist()
list_of_all_titles
```

```python
#finding the close match the movie name given by the user
find_close_match=difflib.get_close_matches(movie_name,list_of_all_titles)
find_close_match
```

```python
close_match=find_close_match[0] #index 
close_match
```

```python
#finding the index of movie based on title
index_of_the_movie=movies_data[movies_data.title==close_match]['index'].values[0]
index_of_the_movie
```

```python
#getting a list of similar movie
similarity_score=list(enumerate(similarity[index_of_the_movie]))
similarity_score
```

```python
len(similarity_score)
```

```python
#we need to sort based on higher similarity and lower similarity

sorted_similar_movies=sorted(similarity_score,key=lambda x:x[1],reverse=True)
sorted_similar_movies
```

```python
#print the name of the similar movies based on the index
print('Movie suggested for you')
i=1
for movie in sorted_similar_movies:
    index=movie[0]
    title_from_index=movies_data[movies_data.index==index]['title'].values[0]
    if(i<10):
        print(i,'-- ',title_from_index)
        i+=2
```

```python
#getting the movie name from the user
director_name=input('Enter the Director name:')

list_of_all_titles=movies_data['director'].tolist()

find_close_match=difflib.get_close_matches(director_name,list_of_all_titles)

close_match=find_close_match[0]

index_of_the_movie=movies_data[movies_data.director==close_match]['index'].values[0]

similarity_score=list(enumerate(similarity[index_of_the_movie]))

sorted_similar_movies=sorted(similarity_score,key=lambda x:x[1],reverse=True)


print('Movie suggested for you: ')

i=1
for movie in sorted_similar_movies:
    index=movie[0]
    title_from_index=movies_data[movies_data.index==index]['title'].values[0]
    if(i<30):
        print(i,'-- ',title_from_index)
        i+=3
```
